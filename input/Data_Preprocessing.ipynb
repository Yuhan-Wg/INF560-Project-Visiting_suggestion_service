```python
import gzip
```

```python
import numpy as np
import pandas as pd
```

```python
num_NaN = 0
for num in range(10):
    with gzip.open('input0'+str(num)+'.gz', 'rb') as f:
        f.readline()
        data = f.read()
        for x in data:
            if x is np.nan or x != x:
                num_NaN += 1
```

```python
print(num_NaN)
```

```python
for num in range(10,100):
    with gzip.open('input'+str(num)+'.gz', 'rb') as f:
        f.readline()
        data = f.read()
        for x in data:
            if x is np.nan or x != x:
                num_NaN += 1
```

```python
print(num_NaN)
```
