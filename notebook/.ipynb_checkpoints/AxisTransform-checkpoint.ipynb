```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import gc
```

```python
df = pd.read_csv("../input/Merged_Input.csv", usecols=['Level', 'lng','lat'])
```

```python
df.head()
```

```python
mask=(df.Level == 'Level B1')
```

```python
df.Level.unique()
```

```python
%%time
plt.figure(figsize=(10,6))
plt.scatter(df[mask].head(10000000).lng, df[mask].head(10000000).lat,alpha=0.01,s=2)
plt.xlim([df[mask].head(10000000).lng.min(),df[mask].head(10000000).lng.max()])
plt.ylim([df[mask].head(10000000).lat.min(),df[mask].head(10000000).lat.max()])
plt.show()
```

```python
#B1
print(df.loc[mask].lng.max()-df.loc[mask].lng.min())
print(df.loc[mask].lat.max()-df.loc[mask].lat.min())
print(df.loc[mask].lng.min())
print(df.loc[mask].lat.min())
```

```python
#1
print(df.loc[mask].lng.max()-df.loc[mask].lng.min())
print(df.loc[mask].lat.max()-df.loc[mask].lat.min())
print(df.loc[mask].lng.min())
print(df.loc[mask].lat.min())
```

```python
%%time
lngRange = 0.0013678100000049653 #df.loc[mask].lng.max()-df.loc[mask].lng.min()
latRange = 0.0015226099999998688 #df.loc[mask].lat.max()-df.loc[mask].lat.min()
tanTheta = latRange/lngRange
cosTheta = 1 / np.sqrt(1 + tanTheta ** 2)
sinTheta = tanTheta * cosTheta
lngMin = 103.850815 #df.loc[mask].lng.min()
latMin = 1.28961633 #df.loc[mask].lat.min()

lngMinimum, lngMaximum = float("inf"), float("-inf")
latMinimum, latMaximum = float("inf"), float("-inf")

for i,chunk in enumerate(pd.read_csv("../input/Merged_Input.csv", usecols=['Level', 'lng','lat'],chunksize=260*10**6 / 10)):
    print("chunk %d"%i)
    mask=(chunk.Level == 'Level 1')
    chunk.loc[mask,'lng'] = chunk.loc[mask,'lng']-lngMin
    chunk.loc[mask,'lat'] = chunk.loc[mask,'lat']-latMin# -latMax/lngMax * df.loc[mask,'transformedLng']
    chunk.loc[mask,'lng'], chunk.loc[mask,'lat']=chunk.loc[mask,'lng'] * cosTheta + chunk.loc[mask,'lat'] * sinTheta, chunk.loc[mask,'lat'] * cosTheta - chunk.loc[mask,'lng'] * sinTheta
    if len(chunk.loc[mask])>0:
        lngMinimum, lngMaximum = min(lngMinimum, min(chunk.loc[mask,'lng'])), max(lngMaximum, max(chunk.loc[mask,'lng']))
        latMinimum, latMaximum = min(latMinimum, min(chunk.loc[mask,'lat'])), max(latMaximum, max(chunk.loc[mask,'lat']))
        print(lngMinimum, lngMaximum)
        print(latMinimum, latMaximum)

print("\n\n\n\n\n")
print(lngMinimum, lngMaximum)
print(latMinimum, latMaximum)
gc.collect()
```
